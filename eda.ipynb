### Analysis of AMEO Data

#### Dataset Information

<h5> Aspiring Minds shared the dataset from the 2015 Aspiring Mind Employment Outcome (AMEO). Students majoring in engineering are the only ones primarily restricted in this study. The dataset includes the standardised test results from three distinct domains (cognitive capabilities, technical skills, and personality skills) as well as the career outcomes of engineering graduates as dependent variables (salary, job titles, and job locations). Demographic features are also included in the dataset. The dataset has four thousand data points and about forty independent factors. The nature of the independent variables is both continuous and categorical. Every candidate in the dataset has a unique identifier.

#### Agenda

This exploratory data analysis's (EDA) objective is to thoroughly examine the dataset that has been provided, with a focus on figuring out how different variables relate to the target variable, salary.

#### Converting dtypes of DOL,DOJ,DOB to datetime as they are in object dtype

```python
import numpy as np 
df['10board'] = df['10board'].replace({'0':np.nan})
df['12board'] = df['12board'].replace({'0':np.nan})
df['GraduationYear'] = df['GraduationYear'].replace({0:np.nan})
df['JobCity'] = df['JobCity'].replace({'-1':np.nan})
df['Domain'] = df['Domain'].replace({-1:np.nan})
df['ElectronicsAndSemicon'] = df['ElectronicsAndSemicon'].replace({-1:0})
df['ComputerScience'] = df['ComputerScience'].replace({-1:0})
df['ComputerProgramming'] = df['ComputerProgramming'].replace({-1:np.nan})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Define numerical features
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']

# Calculate the number of rows and columns for subplots
num_rows = (len(numeric_features) + 2) // 3  # Assuming 3 columns per row
num_cols = min(3, len(numeric_features))

# Set the overall size of the figure
plt.figure(figsize=(15, 4 * num_rows))

# Loop through each numerical feature for boxplots
for i, column in enumerate(numeric_features, 1):
    try:
        # Create subplots
        plt.subplot(num_rows, num_cols, i)

        # Box Plot for the current numerical feature
        sns.boxplot(x=df[column], color='skyblue')
        plt.title(f'Boxplot - {column}')

        # Add mean as a red dot for reference
        plt.plot(df[column].mean(), marker='o', markersize=8, color='red', label='Mean')

        # Add median as a green dot for reference
        plt.plot(df[column].median(), marker='o', markersize=8, color='green', label='Median')

        # Add legend
        plt.legend()
    except TypeError as e:
        print(f"Error in column: {column}")
        print(f"Error message: {e}")

# Adjust layout to prevent overlapping subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Define a function to create countplots with top N unique values
def plot_top_n_countplots(df, categorical_column, top_n=15):
    top_values = df[categorical_column].value_counts().head(top_n).index
    filtered_df = df[df[categorical_column].isin(top_values)]

    plt.figure(figsize=(8, 5))
    sns.countplot(x=filtered_df[categorical_column], palette='viridis')
    plt.title(f'Top {top_n} Values - {categorical_column}')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Define categorical features
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# Plot countplots for the top N unique values in each categorical column
for column in categorical_features:
    plot_top_n_countplots(df, column, top_n=15)
```

```python
import pandas as pd

df = pd.read_csv('data/amcat.csv')

# Testing the claim about earnings
earnings_claim = df[(df['Specialization'].isin(['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer'])) & (df['Salary'] >= 2.5) & (df['Salary'] <= 3.0)]

# Relationship between gender and specialization
gender_specialization_relationship = df.groupby(['Gender', 'Specialization']).size().unstack()

earnings_claim, gender_specialization_relationship
```

#### The analysis has been completed:

Testing the Claim about Earnings:

The data does not directly support the claim made in the Times of India article. There are no fresh graduates in the dataset with earnings between 2.5-3 lakhs specifically in the roles of Programming Analyst, Software Engineer, Hardware Engineer, or Associate Engineer.

```python
# Taking the top 15 specializations based on the number of fresh graduates
top_15_specializations = df['Specialization'].value_counts().nlargest(15).index

# Filtering the dataframe for the top 15 specializations
earnings_distribution_top_15 = df[df['Specialization'].isin(top_15_specializations)]

# Visualizing the distribution of earnings for fresh graduates in the top 15 specializations
plt.figure(figsize=(12, 6))
earnings_distribution_top_15['Specialization'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Earnings for Fresh Graduates in Top 15 Specializations')
plt.xlabel('Specialization')
plt.ylabel('Number of Fresh Graduates')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### visualization showing the distribution of earnings for fresh graduates in the top 15 specializations.

#### Research Questions
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
import matplotlib.pyplot as plt

# Visualizing the relationship between gender and specialization for the top 15 specializations
top_15_specializations = df['Specialization'].value_counts().head(15).index

plt.figure(figsize=(12, 8))
for spec in top_15_specializations:
    subset = df[df['Specialization'] == spec]
    subset['Gender'].value_counts().plot(kind='bar', color=['blue', 'orange'])
    plt.title(f'Gender Distribution for {spec}')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.legend(['Male', 'Female'])
    plt.show()
```

#### Mostly majority are men in every specialization
